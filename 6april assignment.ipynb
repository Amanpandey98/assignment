Q1. What is the mathematical formula for a linear SVM?

A linear support vector machine (SVM) is a type of binary classification algorithm that finds a linear boundary that separates two classes of data. The mathematical formula for a linear SVM can be expressed as:

f(x) = sign(w^T x + b)

where:

f(x) is the decision function that outputs either +1 or -1 for a given input vector x
w is the weight vector that defines the direction of the separating hyperplane
b is the bias term that shifts the hyperplane away from the origin
x is the input vector of features
T denotes the transpose of a matrix.

Q2. What is the objective function of a linear SVM?

The objective function of a linear support vector machine (SVM) is to find the hyperplane that best separates the data into two classes. In particular, the objective function of a linear SVM aims to maximize the margin between the two classes while minimizing the classification error.

Q3. What is the kernel trick in SVM?

The kernel trick in support vector machines (SVM) is a technique that allows a linear algorithm to efficiently solve nonlinear problems by transforming the input data into a higher-dimensional space.

Q4. What is the role of support vectors in SVM Explain with example

In support vector machines (SVM), support vectors are the data points that lie closest to the decision boundary (i.e., the hyperplane) and have a non-zero weight in defining the decision boundary. These support vectors play a crucial role in SVMs because they determine the position and orientation of the decision boundary, as well as the size of the margin.

The support vectors are chosen by the SVM algorithm during training, and they are the only data points that contribute to defining the decision boundary. This makes SVMs very efficient in high-dimensional feature spaces, as only a subset of the data points needs to be considered for the classification task.

Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in
SVM?

1. Hyperplane: In SVM, the hyperplane is the decision boundary that separates the data points into two classes. It is a linear function that can be represented as w^T x + b = 0, where w is the weight vector, x is the input vector of features, and b is the bias term. The hyperplane can be thought of as a plane that divides the feature space into two parts, where each part corresponds to one of the classes.

2. Marginal plane: The marginal plane is the hyperplane that is equidistant from the support vectors of the two classes. It is the plane that runs parallel to the hyperplane and is located on the boundary of the margin. The margin is defined as the region between the hyperplane and the two marginal planes. The distance between the hyperplane and the marginal plane is called the margin width.

3. Soft margin: In some cases, the data points may not be linearly separable, and it may not be possible to find a hyperplane that separates the two classes without any errors. In such cases, we can allow some misclassification errors and find a hyperplane that separates the data points with the fewest errors possible. This is known as soft margin SVM.

4. Hard margin: In some cases, the data points are linearly separable, and it is possible to find a hyperplane that perfectly separates the two classes without any errors. This is known as hard margin SVM.



Q6. SVM Implementation through Iris dataset.

Bonus task: Implement a linear SVM classifier from scratch using Python and compare its
performance with the scikit-learn implementation.
~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
~ Compute the accuracy of the model on the testing setl
~ Plot the decision boundaries of the trained model using two of the featuresl
~ Try different values of the regularisation parameter C and see how it affects the performance of
the model.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

iris = load_iris()
X = iris.data[:, :2]  # we only take the first two features for visualization purposes
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
clf = LinearSVC()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
# create a mesh to plot in
h = 0.02  # step size in the mesh
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# plot the decision boundary
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, cmap=plt.cm.Accent, alpha=0.8)

# plot the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())
plt.show()
C_values = [0.1, 1, 10, 100]
for C in C_values:
    clf = LinearSVC(C=C)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("C =", C, "Accuracy:", accuracy)

```
