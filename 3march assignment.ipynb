```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
The five plots that we can plot using the Seaborn library along with their uses:

1.Scatterplot: A scatterplot displays the relationship between two variables by plotting them as a set of points on a two-dimensional plane. It is useful for identifying trends and patterns in data and for visualizing the strength and direction of a correlation between two variables.

2.Lineplot: A lineplot is used to display the trend of a variable over time or across categories. It is useful for visualizing changes in a variable over time and for comparing multiple variables simultaneously.

3.Barplot: A barplot displays the distribution of a categorical variable using bars of different heights. It is useful for comparing the frequency or proportion of different categories and for identifying patterns and trends in the data.

4.Boxplot: A boxplot is used to display the distribution of a continuous variable. It is useful for identifying outliers and for visualizing the spread, central tendency, and skewness of the data.

5.Heatmap: A heatmap displays the relationship between two categorical variables using a color-coded matrix. It is useful for identifying patterns and trends in large datasets and for visualizing the frequency and strength of a relationship between two variables.
```

```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
```

```python
import seaborn as sns
fmri=sns.load_dataset("fmri")
sns.lineplot(x="timepoint",y="signal",hue="event",style="region",data=fmri)
```

```python
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
```

```python
import seaborn as sns
titanic_data=sns.load_dataset("titanic")
sns.boxplot(x="pclass",y="age",data=titanic_data)
sns.boxplot(x="pclass",y="fare",data=titanic_data)

```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
import seaborn as sns
diamonds_data=sns.load_dataset("diamonds")
sns.histplot(x="price",hue="cut",data=diamonds_data,kde=True)
```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
```

```python
import seaborn as sns
iris_data=sns.load_dataset("iris")
sns.pairplot(hue="species",data=iris_data)
```

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
import seaborn as sns

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix for the heatmap
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot a heatmap for the flights data
sns.heatmap(data=flights_pivot, cmap="Blues")
```
