```python
Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps
with the highest number of installs.
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("googleplaystore.csv")
# Remove duplicates
df.drop_duplicates(subset="App", keep="last", inplace=True)
# Convert the "Installs" column to numeric
df["Installs"] = df["Installs"].str.replace(",", "")
df["Installs"] = df["Installs"].str.replace("+", "")
df["Installs"] = pd.to_numeric(df["Installs"])
# Create a frequency table of the top 10 most popular apps
top_apps = df.sort_values(by="Installs", ascending=False).head(10)
print(top_apps[["App", "Installs"]])
```

Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the
Google Playstore dataset to identify the most popular app categories. How would you approach this
task, and what features would you analyze to make recommendations to the company?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("googleplaystore.csv")
# Remove duplicates
df.drop_duplicates(subset="App", keep="last", inplace=True)
# Convert the "Installs" column to numeric
df["Installs"] = df["Installs"].str.replace(",", "")
df["Installs"] = df["Installs"].str.replace("+", "")
df["Installs"] = pd.to_numeric(df["Installs"])
# Compute the total number of installs for each category
category_installs = df.groupby("Category")["Installs"].sum().reset_index()

# Sort the categories by the number of installs in descending order
category_installs = category_installs.sort_values(by="Installs", ascending=False)

# Print the top 10 most popular categories and their number of installs
print(category_installs.head(10))
```

Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the
most successful app developers. What features would you analyze to make recommendations to the
company, and what data visualizations would you use to present your findings?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("googleplaystore.csv")

# Clean the dataset by removing duplicates and invalid data
df.drop_duplicates(subset="App", keep="last", inplace=True)

# Compute the number of apps developed by each developer
apps_per_developer = df.groupby("Developer")["App"].count().reset_index()

# Sort the developers by the number of apps they have developed in descending order
apps_per_developer = apps_per_developer.sort_values(by="App", ascending=False)

# Create a bar chart to show the number of apps developed by each developer
plt.figure(figsize=(12, 6))
plt.bar(apps_per_developer["Developer"].head(10), apps_per_developer["App"].head(10))
plt.title("Top 10 Developers by Number of Apps")
plt.xlabel("Developer")
plt.ylabel("Number of Apps")
plt.xticks(rotation=45)
plt.show()

# Compute the average rating and total number of installs for each developer's apps
ratings_installs_per_developer = df.groupby("Developer").agg({"Rating": "mean", "Installs": "sum"}).reset_index()

# Create a scatter plot to show the relationship between the average rating and total number of installs for each developer's apps
plt.figure(figsize=(8, 6))
plt.scatter(ratings_installs_per_developer["Rating"], ratings_installs_per_developer["Installs"])
plt.title("Developer Success by Ratings and Installs")
plt.xlabel("Average Rating")
plt.ylabel("Total Installs")
plt.show()

# Create a box plot to show the distribution of average prices for each developer's apps
prices_per_developer = df.groupby("Developer")["Price"].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.boxplot(prices_per_developer["Price"])
plt.title("Distribution of Average Prices by Developer")
plt.ylabel("Average Price")
plt.show()
```

Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to
launch a new app. What features would you analyze to make recommendations to the company, and
what data visualizations would you use to present your findings?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("googleplaystore.csv")

# Clean the dataset by removing duplicates and invalid data
df.drop_duplicates(subset="App", keep="last", inplace=True)

# Convert the "Last Updated" column to a datetime format
df["Last Updated"] = pd.to_datetime(df["Last Updated"])

# Extract the month and year from the "Last Updated" column
df["Month"] = df["Last Updated"].dt.strftime("%Y-%m")

# Compute the number of app releases per month
apps_per_month = df.groupby("Month")["App"].count().reset_index()

# Create a line chart to show the trend of the number of app releases by month
plt.figure(figsize=(12, 6))
plt.plot(apps_per_month["Month"], apps_per_month["App"])
plt.title("Number of App Releases by Month")
plt.xlabel("Month")
plt.ylabel("Number of App Releases")
plt.xticks(rotation=45)
plt.show()

# Compute the average rating and total number of installs of app releases by month
ratings_installs_per_month = df.groupby("Month").agg({"Rating": "mean", "Installs": "sum"}).reset_index()

# Create a bar chart to show the average rating and total number of installs of app releases by month
plt.figure(figsize=(12, 6))
plt.bar(ratings_installs_per_month["Month"], ratings_installs_per_month["Rating"], alpha=0.5, label="Average Rating")
plt.bar(ratings_installs_per_month["Month"], ratings_installs_per_month["Installs"], alpha=0.5, label="Total Installs")
plt.title("App Success by Month")
plt.xlabel("Month")
plt.ylabel("Rating/Installs")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
